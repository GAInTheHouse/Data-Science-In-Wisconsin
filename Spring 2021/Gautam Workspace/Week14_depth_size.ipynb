```python
import pandas as pd
from math import sin, cos, asin, sqrt, pi
import matplotlib
import geopandas
from matplotlib import pyplot as plt
import math
import numpy as np
```

```python
data=geopandas.read_file("zip://../dataset/Water_Main_Breaks-shp.zip")
data
```

```python
data["FacilityID"]=data["MainID"]
```

```python
wm = geopandas.read_file("zip://../dataset/WaterMains.zip")
wm
```

```python
def cdf_calculator(df):
    prev=df.index[0]
    for i in df.index:
        if i !=prev:
            df["Material"][i]+=df["Material"][prev]
        prev=i
    df["Material"]/=max(df["Material"])
    return df
```

```python
size_material=wm[["Material","Diameter"]]
size_material=size_material.sort_values("Diameter").reset_index().dropna()
size_material=size_material[["Material","Diameter"]]
```

```python
material_list_size=wm["Material"].unique()[:-1]
material_list_size=np.append(material_list_size,"Total")
material_list_size
```

```python
total_cdf=cdf_calculator(size_material.groupby(by='Diameter').count())
di_cdf=cdf_calculator(size_material[size_material["Material"]=="DI"].groupby(by='Diameter').count())
ci_cdf=cdf_calculator(size_material[size_material["Material"]=="CI"].groupby(by='Diameter').count())
copper_cdf=cdf_calculator(size_material[size_material["Material"]=="COPPER"].groupby(by='Diameter').count())
sand_cdf=cdf_calculator(size_material[size_material["Material"]=="SAND"].groupby(by='Diameter').count())
spun_cdf=cdf_calculator(size_material[size_material["Material"]=="SPUN"].groupby(by='Diameter').count())
hdpe_cdf=cdf_calculator(size_material[size_material["Material"]=="HDPE"].groupby(by='Diameter').count())
cipp_cdf=cdf_calculator(size_material[size_material["Material"]=="CIPP"].groupby(by='Diameter').count())
pvc_cdf=cdf_calculator(size_material[size_material["Material"]=="PVC"].groupby(by='Diameter').count())
```

```python
matplotlib.rcParams.update({'font.size': 12})
fig,ax1 = plt.subplots(figsize=(6,4))
di_cdf.plot(color="cyan",ax=ax1)
ci_cdf.plot(color="blue",ax=ax1)
copper_cdf.plot(color="yellow",ax=ax1)
sand_cdf.plot(color="brown",ax=ax1)
spun_cdf.plot(color="green",ax=ax1)
hdpe_cdf.plot(color="black",ax=ax1)
cipp_cdf.plot(color="violet",ax=ax1)
pvc_cdf.plot(color="pink",ax=ax1)
total_cdf.plot(color="r",ax=ax1)
ax1.legend(material_list_size)
ax1.set_xlim([0.1,25])
ax1.set_ylim([0,1.1])
ax1.set_ylabel("Cummulative % of All Pipes")
ax1.set_xlabel("Pipe Size (inches)")
ax1.spines["top"].set_visible(False)
ax1.spines["right"].set_visible(False)
plt.tight_layout()
#plt.savefig('depth_cdf.pdf',format='pdf',bbox_inches="tight")
```
