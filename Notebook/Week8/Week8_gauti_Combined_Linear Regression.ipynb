```python
import csv
import numpy
reader = csv.reader(open("Jan_Breaks.csv", "r"), delimiter=",")
x = list(reader)

result = numpy.array(x[1:][:]).astype("float")
```

```python
X=result[:,1:7]
y=result[:,7]
print(y.shape)
y=y.reshape(357,1)
print(X,y)
```

```python
w=numpy.linalg.inv(X.transpose()@X)@X.transpose()@y
w
```

```python
import csv
import numpy
reader = csv.reader(open("test.csv", "r"), delimiter=",")
x = list(reader)

test = numpy.array(x[1:][:]).astype("float")
test=test[:,1:]
print(test)
```
