```python
import csv
import numpy
from numpy import genfromtxt
my_data = genfromtxt('dataset/training_matrixv2.csv', delimiter=',')
train = my_data[1:,1:]
```

```python
training_x=[]
training_y=[]
for i in range(1,13):
    mat=train.copy()
    mat=mat[mat[:,10] == i]
    mat=mat[:,:10]
    training_x.append(mat[:,:9])
    training_y.append(mat[:,9])

training_x[0][0]
```

```python
w=[]
for i in range(12):
    X=training_x[i]
    y=training_y[i]
    w.append(numpy.linalg.inv(X.transpose()@X)@X.transpose()@y)
w
```

```python
test_data = genfromtxt('dataset/test.csv', delimiter=',')
test = test_data[1:,1:]
```

```python
test_x=[]
for i in range(1,13):
    mat=test.copy()
    mat=mat[mat[:,9] == i]
    mat=mat[:,:9]
    test_x.append(mat)

test_x[0][0]
```

```python
result=[]
prediction=[]
for i in range(0,12):
    X=test_x[i]
    weight=w[i]
    y=X@weight
    y = numpy.reshape(y,(y.size,1))
    X=numpy.hstack((X,y))
    m,n=X.shape
    imatrix=numpy.full((m,1),i+1,dtype=int)
    prediction.append(numpy.hstack((X,imatrix)))
prediction[0]
   
```

```python
plotmatrix=prediction[0]
for i in range(1,12):
    #plotmatrix=plotmatrix.append(plotmatrix,prediction[1])
    plotmatrix=numpy.vstack((plotmatrix,prediction[i]))

plotmatrix=plotmatrix[:,(2,5,9,10)]
plotmatrix
```

```python
import pandas as pd
column_values=["pipe_depth","pipe_size","Number","Month"]
df = pd.DataFrame(data = plotmatrix,    
                  columns = column_values)
df.to_csv('manupilationv2.csv')
```
