```python
import pandas as pd
from math import sin, cos, asin, sqrt, pi
import matplotlib
import geopandas
from matplotlib import pyplot as plt
import numpy
import math
import csv
from pandas import Series
from pandas import DataFrame
%matplotlib inline
```

```python
def process_csv(filename):
    exampleFile = open(filename, encoding="utf-8")
    exampleReader = csv.reader(exampleFile)
    exampleData = list(exampleReader)
    exampleFile.close()
    return exampleData

# use process_csv to pull out the header and data rows
csv_rows = process_csv("water_main_breaks.csv")
csv_header = csv_rows[0]
csv_data = csv_rows[1:]
```

```python
from collections import OrderedDict 
pipe_size={}
index= csv_header.index("pipe_size")

for i in csv_data:
    if i[index] in pipe_size:
        pipe_size[i[index]]+=1
    else:
        pipe_size[i[index]]=1

pipe_size.pop('')
pipe_size
```

```python
df=DataFrame(Series(pipe_size))
df.index=df.index.astype(float)
df=df.sort_index()
df=df.rename(columns={0:'Main Breaks'})
df
```

```python
ax1= df.plot.line(color="r",xticks=[0,2,4,6,8,10,12,14,16],figsize=[15,10])
ax1.set_xlim([0,17])
ax1.set_ylim([0,2500])
ax1.set_ylabel("Number of Pipes")
ax1.set_xlabel("Pipe Size (ft.)")
ax1.spines["top"].set_visible(False)
ax1.spines["right"].set_visible(False)
```
